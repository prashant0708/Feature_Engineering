###### InvoiceDate is in datetime only.

```python
## extract the time  apart 

df['time']=df["InvoiceDate"].dt.time
```

```python
df.head(2)
```

```python
## Extract the hours apart

df['hours']=df['InvoiceDate'].dt.hour

## Extract the minutes apart

df['minutes']=df['InvoiceDate'].dt.minute

## Extract the second apart

df['Second']=df['InvoiceDate'].dt.second
```

```python
df.head()
```

```python
## extract microsecond and nenosecond apart

df['miscrosecond']= df['InvoiceDate'].dt.microsecond
df['nenosecond']=df['InvoiceDate'].dt.nanosecond
```

## Unsing the Feature Engine Library to extract the data

```python
pip install feature_engine
```

```python
## import feature engine

from feature_engine.datetime import DatetimeFeatures
```

```python
df1=df.iloc[:,0:8]
df1.head()
```

## Extract all possible feature using Feature Engine

```python
## intilize the date time feature 
dft=DatetimeFeatures(
variables=None,
 features_to_extract="all"   
    
    
)
```

```python
dft.fit(df1)
```

```python
df2=dft.transform(df1)
```

```python
df2.head(2)
```

```python
df2.filter(like='InvoiceDate')
```

## Extract the Most Common feature 

```python
dfts=DatetimeFeatures(variables=None,
 features_to_extract=None 
)
```

```python
dfts.fit(df1)
df3=dfts.transform(df1)
```

```python
df3.filter(like="InvoiceDate")
```

## Extract the User Define 

```python
dtfs_1=DatetimeFeatures(variables=None,features_to_extract=['month','year','day_of_month','week'])

dtfs_1.fit(df1)

df4=dtfs_1.transform(df1)
```
